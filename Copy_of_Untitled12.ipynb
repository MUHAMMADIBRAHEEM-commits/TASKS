```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
data = pd.read_csv("/content/iris.data.csv")
```

```python
sns.histplot(data["1.4"], kde=True)
plt.title("Distribution of 1.4")

# Calculate mean and median
val_mean = data["1.4"].mean()
val_median = data["1.4"].median()

print("Mean:", val_mean)
print("Median:", val_median)

# Add vertical lines for mean and median
plt.axvline(val_mean, color="red", linestyle="--", label="Mean")
plt.axvline(val_median, color="green", linestyle="-", label="Median")

plt.legend()
plt.show()
```

```python
# Histogram for "3.5"
sns.histplot(data["3.5"], kde=True)
plt.title("Distribution of 3.5")

# Calculate mean and median
val_mean = data["3.5"].mean()
val_median = data["3.5"].median()

print("Mean:", val_mean)
print("Median:", val_median)

# Add vertical lines for mean and median
plt.axvline(val_mean, color="red", linestyle="--", label="Mean")
plt.axvline(val_median, color="green", linestyle="-", label="Median")

plt.legend()
plt.show()
```
